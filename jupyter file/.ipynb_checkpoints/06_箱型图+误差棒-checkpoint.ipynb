# 箱形图

最小值，下四分位数，中位数，上四分位数，最大值;(异常值)能够识别异常值判断数据偏态，尾重;质量管理


## boxplot（）  
boxplot(x,sym=None,vert=None,whis=Non,positions=None,  
widths=None,patch_artist=None,meanline=None,showcaps=None,  
showbox=None,showfliers=None,labels=None,boxprops=None)  
×:箱型图的数据  
sym:异常值的符号，默认空心圆圈  
vert:表示是否将箱型图进行垂直摆放，默认垂直摆放whis:表示与上下四分位的距离，默认1.5倍的四分位差positions:箱体的位置  
widths:箱体宽度，默认0.5  
patch_artist:是否填充箱体颜色，默认不填充  
meanline:是否用横跨箱体的线条标识出中位数，默认不使用  
showcaps:是否显示箱体顶部和底部的横线，默认显示  
showbox:是否显示箱型图的箱体，默认显示  
showfliers:是否显示异常值，默认显示  
labels:箱型图的标签  
boxprops:控制箱型图属性的字典  



```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(100)
data2 = np.random.randn(200)
plt.boxplot([data, data2], meanline=True, widths=0.3, patch_artist=True, showfliers=False)

plt.show()
```

### 实例 2017和2018全国发点统计
两列数据作为x轴,2018年和2017年作为y轴刻度标签：绘制箱型图

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.array([5200, 5254.2, 5283.4, 5107.8, 5443.3, 5550.6, 6400.2, 6404.9, 5483.1, 5330.2, 5543, 6199.9])
data2 = np.array([4605.2, 4710.3, 5168.9, 4767.2, 4947, 5203, 6047.4, 5945.5, 5219.6, 5038.1, 5196.3, 5698.6])
sj = ([2017, 2018])
plt.boxplot([data, data2], meanline=True, widths=0.5, patch_artist=True, vert=False, labels=sj)

plt.show()
```

# 误差棒图

## 使用errorbar（）绘制误差棒图

errorbar(x, y, xerr=None, yerr=None,fmt='', ecolor=None, elinewidth=None, capsize=None, capthick=None)  
X，y:表示数据点的位置。  
xerr，yerr:表示数据的误差范围。  
lfmt:表示数据点的标记样式和数据点之间连接线的样式。  
Iecolor:表示误差棒线条的颜色  
elinewidth:表示误差棒的线条宽度。  
capsize:表示误差棒边界横杆的大小。  
capthick:表示误差棒边界横杆的厚度。  


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(5)
y = (25, 32, 34, 20, 25)
y_offset = (3, 5, 2, 3, 3)
plt.errorbar(x, y, yerr=y_offset, capsize=3, capthick=2)
plt.show()
```

```python
x = np.arange(3)
y1 = np.array([2.04, 1.57, 1.63])
y2 = np.array([1.69, 1.61, 1.64])
y3 = np.array([4.65, 4.99, 4.94])
y4 = np.array([3.93, 2.33, 4.10])
y1s = (0.16, 0.08, 0.10)
y2s = (0.27, 0.14, 0.14)
y3s = (0.34, 0.32, 0.29)
y4s = (0.23, 0.23, 0.39)
bar_width = 0.2
wu = dict(capsize=3, capthick=3)
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.bar(x, y1, tick_label=['春季', '夏季', '秋季'], width=bar_width, yerr=y1s, error_kw=wu)
plt.bar(x + bar_width, y2, width=bar_width, yerr=y2s, error_kw=wu)
plt.bar(x + bar_width + bar_width, y3, width=bar_width, yerr=y3s, error_kw=wu)
plt.bar(x + bar_width + bar_width + bar_width, y4, width=bar_width, yerr=y4s, error_kw=wu)
plt.show()
```
