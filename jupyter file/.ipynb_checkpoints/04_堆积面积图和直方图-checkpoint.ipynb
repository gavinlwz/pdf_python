# 1. 堆积面积图


使用stackplot()绘制堆积面积图  
stackplot(x, y, labels=(), baseline='zero’）  
x :x轴数据，可以是一维数组  
y:y轴数据，可以是二维数组或一维数组  
labels:每组折线以及填充区域的标签  
baseline:计算基线的方法，'zero'表示恒定零基线，'sym对称于零基线,'wiggle'表示最小平方斜率的总和，'weighted_wiggle'用于说明每层的大小



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(6)
y1 = np.array([1, 2, 5, 6, 4, 7])#图一
y2 = np.array([2, 3, 4, 5, 6, 7])#图二
y3 = np.array([3, 4, 3, 8, 5, 6])#图三
#堆积面积图
plt.stackplot(x, y1, y2, y3)
plt.show()
```

## 实例：


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,13)
y1 = np.array([198, 215, 245, 222, 200, 236, 201, 253, 236, 200, 266, 290])
y2 = np.array([203, 236, 200, 236, 269, 216, 298, 333, 301, 349, 360, 358])
y3 = np.array([185, 205, 226, 199, 238, 200, 250, 209, 246, 219, 253, 288])
plt.stackplot(x, y1, y2, y3)
plt.show()
```

# 直方图 

使用hist()绘制直方图
hist(x, bins=None, range=None, cumulative=False, histtype=' bar',align='mid' , orientation='vertical' , rwidth=0, stacked=False)    
x:x轴数据，可以是单个或多个数组  
bins:表示矩形条个数，默认10个  
range:表示数据范围，如果不填，范围是(x.min O , x.max())  
cumulative:表示是否计算累计频数或频率  
histtype:表示直方图类型,'bar'传统直方图，'barstacked'堆积直方图，'step'未填充线条的直方图， 'setpfilled'填充的线条直方图  
align:矩形条边界的对其方式， 'left', 'right' , 'mid'，默认是mid  
rwidth:矩形条的宽度百分比，默认0;如果histtype值是'step'或'stepfilled',忽略rwidth参数  
stacked:矩形条是否以堆积的形式摆放  


概念:直方图用来展示数据分布情况:先对数据进行分组，再用面积表示各组频数的多少，矩形的高度代表每一组的频数或者频率，宽度表示组距。数据有连续性，直方图的各个矩形通常是连续排列的


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num = np.random.randint(0, 100, 50)
plt.hist(num, bins=8, histtype='stepfilled')
plt.show()
```
