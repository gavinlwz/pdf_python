# 饼图+圆环图

### 使用pei（）绘制1饼图和圆环图

pie(x, explode=None, labels=None, autopct=None, pctdistance=0.6, shadow=False, labeldistance=1.1,startangle=None, radius=None,   wedgreprops=None,textprops=None, center=(0,0), frame=False)  
x:表示扇形或楔形的数据  
explode:表示扇形或楔形离开圆心的距离  
labels:扇形或楔形对应的文本标签   
autopct:控制扇形或楔形数值显示的字符串，可以指定小数点位数  
pctdistance:扇形或楔形对应的数值标签距离圆心的比例，默认0.6  
shadow:是否显示阴影  
labeldistance:文本标签的绘制位置（相对于半径的比例），默认1.1  
startangle:表示起始绘制的角度，默认从x轴正方向逆时针绘制  
radius:表示扇形或楔形的半径  
wedgeprops:控制扇形或楔形属性的字典;wedgeprops={'width’:0.6}  
textprops:表示图表文本属性的字典  
center:图表中心点位置，默认(0,0)  
frame:是否显示图框  


```python
import numpy as np
import matplotlib.pyplot as plt
data = np.array([20, 60, 50, 12, 30, 55])
pie_labels = np.array(['A', 'B', 'C', 'D', 'E', 'F'])
plt.pie(data, radius=1.5, labels=pie_labels, autopct='%3.1f%%')
plt.show()
```

# 散点图+气泡图

scatter (x, y,s=None, c=None, marker=None,cmap=None, norm=None, vmin=None,  
vmax=None,alpha=None,linewidths=None, edgecolors=None)  
x, y:表示数据点位置  
s:数据点大小  
c:表示数据点颜色  
marker:数据点样式，默认圆形  
cmap:数据点颜色映射，仅当参数c为浮点数组时才能使用  
norm:表示数据亮度，取值为01~1  
vmin, vmax:亮度最小值，最大值;如果传入norm，则忽略此参数  
alpha:透明度0~1  
linewidths:数据点边缘的宽度  
edgecolors:数据点边缘的颜色  


```python
import numpy as np
import matplotlib.pyplot as plt
num = 50
x = np.random.rand(num)
y = np.random.rand(num)
area = (np.random.rand(num) * 30) ** 2
plt.scatter(x, y, s=area)
plt.show()
```
