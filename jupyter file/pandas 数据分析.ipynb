#  pandas 数据处理

1 series 

```python
s = pd.Series([29, 30, 1, -29], index=["a", "b", "c", "d" ])
print(s.values)  # 查看元素
print(s.index)  # 查看标签
```

```python
# 索引
print(s[3])
print(s["c"])
```

```python
# 标签获取数据
print(s[0:2])
print(s[["b", "c"]])
print(s["b":"c"])
```

```python
# 修改元素
s[1] = -30
s["a"] = 50
print(s)
```

```python
# 追加
s = pd.Series([29, 30, 1, -29], index=["a", "b", "c", "d", ])
n = pd.Series([5], index=["e"])
s.append(n)
print(s)
```

```python
# 删除drop（）
print(s.drop("a"))
print(s.drop(s.index[2]))#索引删除
print(s["c"!=s.index])#以c删除-29
print(s[-29!=s.values])#以-29删除c
```

### 默认升序 排序sort_index()
 ascending=True 升序
 ascending=False降序

```python
s = pd.Series([29, 30, 1, -29], index=["b", "c", "d", "a"])
print("创的数组是\n",s)
print("-----------")
print(s.sort_index())#阿斯克码排序
```

reindex（）对索引重新排序
fill_value= 填充缺省直

```python
# reindex() 
# fill_value=填充值 或空值
s = pd.Series([29, 30, 1, -29], index=["b", "c", "d", "a"])
snew=s.reindex(["a","b","c","d","e"],fill_value=100)
print(snew)
```

### 数据写入
to_csv(filepath,sep='',na_rep='',columns=None,header=True,index=True,index_label=None,mocoding-None)  
 1. filepath,sep='' 文件路径  
 2. sep=''分隔符 string 默认  
 3. na_rep=''缺失值 默认是‘’  
 4. columns=None 接收list，写入得列名，None  
 5. header boolean 表示是否将列名写出默认是 True  
 6. index boolean 表示是否写出行名 默认是 True  
 7. mode 接收string类型 表示数据写入模式w  

```python
import pandas as pd
import numpy as np

data = {
    'color': ['red', 'green', 'gray', 'blue', 'black'],
    'object': ['mug', 'paper', 'desk', 'book', 'pen'],
    'prince': [20, 1.5, 50, 32, 5]
}
df = pd.DataFrame(data)
df.to_csv('files/pandas_csv.csv', sep=',', na_rep='', columns=None, header=True, index=True,
          index_label=None, mode='w', encoding=None)
print(df)
```

## 写入数据库  
to_sql(name,con,if_exists,'fail',index=True,index_label=None,dtype=None)  
1. naem 接收string 类型 数据库表名
2. con 连接数据库
3. if_exists:接收fail‘，’‘replace’；fail表示如果存在 则不执行写入；replace表示如果存在，则将元数据库标删除，在创建，append表示在数据库基础上的再追加数据默认是fail
4. dtype写入数据

```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine("mysql+pymysql://root:root@localhost:3306/d1tk?charset=utf8")
data = {
    'color': ['red', 'green', 'gray', 'blue', 'black'],
    'object': ['mug', 'paper', 'desk', 'book', 'pen'],
    'prince': [20, 1.5, 50, 32, 5]
}
df = pd.DataFrame(data)
df.to_sql("data1", con=engine, if_exists="fail", index=False, index_label="id")
df
```

### 3.缺失值处理（替代）  
fillna（）使用数据对缺失值进行填补  
fillna（mathod="pad"）使用前一个数据进行替换  
fillna（mathod="bfill"）使用后一个数据进行替换  

```python
df = pd.read_csv(r'files\loan_data.csv', header='infer', sep=',',encoding='gbk')
df[df.isnull().values==True]
```

```python
newDF=df.fillna(method="pad")#使用以一个数据进行替代，结构保存在newDF中 后一个数据是bfill
newDF.iloc[6]#切片抽样
```

```python
newDF=df.fillna(method="bfill")#使用以一个数据进行替代，结构保存在newDF中 后一个数据是bfill
newDF.iloc[5]#切片抽样
```
