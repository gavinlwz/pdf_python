### 实例练习 ：天气预报

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(4, 19)  # 生产日期
y_max = ([32, 33, 34, 33, 31, 30, 29, 30, 29, 24, 25, 31, 23, 25, 31])
y_min = ([12, 15, 16, 19, 14, 18, 17, 15, 12, 21, 24, 23, 24, 21, 21])
plt.plot(x, y_max)  # 画最高温
plt.plot(x, y_min)  # 画对低温
plt.show()
#结果：未来天气逐步下降然后反贪去世
```

## 二，绘制柱形图或堆积柱形图

### 2. 使用bar（）绘制柱形图或者堆积柱形图

pyplot是使用bar()绘制柱形图:  
bar (x, height,width=0.8, bottom=None,align=' center' ,tick_label=, xerr=None,yerr=None,error_kw=None)  
x:表示柱形图的x坐标值  
height:表示柱形高度  
width:表示柱形的宽度默认0.8  
bottom:表示柱形底部的y坐标值，默认0  
align:表示柱形的对齐方式 'center'和' edge'， center表示柱形与刻度线居中对齐，edge表示柱形图左边与刻度线对齐
tick_label:表示柱形的刻度标签  
xerr, yerr:表示柱形图添加的水平或者垂直的误差棒  
error_kw:表示误差棒的属性字典；对应errorbar（）的关键字参数  




```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(5)
y1 = np.array([10, 5, 8, 13, 11])
# 宽度
bar_width = 0.3
plt.rcParams['font.sans-serif']=['Simhei']#设置字体
plt.bar(x, y1, tick_label=['小张', '小王', '李雷', '张三', '李四'], width=bar_width)
plt.show()
```

2.多个柱形图

```python
x = np.arange(5)
y1 = np.array([10, 5, 8, 13, 11])
y2 = np.array([10, 2, 15, 18, 19])
# 宽度
bar_width = 0.3
plt.rcParams['font.sans-serif']=['Simhei']#有中文没有可以不设设置字体
plt.bar(x, y1, tick_label=['小张', '小王', '李雷', '张三', '李四'], width=bar_width)
plt.bar(x+bar_width,y2,width=bar_width)
plt.show()
```

```python
x = np.arange(5)
y1 = np.array([10, 5, 8, 13, 11])
y2 = np.array([10, 2, 15, 18, 19])
# 宽度
bar_width = 0.3
plt.rcParams['font.sans-serif'] = ['SimHei']  #有中文没有可以不设设置字体
plt.bar(x, y1, tick_label=['张三', '李四', '王五', '赵六', '孙七'], width=bar_width)#画第一个表并设置名字
plt.bar(x,y2,bottom=y1, width=bar_width)#把第二个数据放到第一个上面
plt.show()
```

### 3. 含有误差棒的柱形图  
xerr，yerr

```python
error = [2, 1, 5, 2, 1]
plt.bar(x, y1, tick_label=['小张', '小王', '李雷', '张三', '李四'], width=bar_width)
plt.bar(x, y2, bottom=y1, width=bar_width, yerr=error)
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt  #绘制图像的模块
import jieba                    #jieba分词

path_txt = 'all.txt'
f = open(path_txt,'r',encoding='UTF-8').read()

# 结巴分词，生成字符串，wordcloud无法直接生成正确的中文词云
cut_text = " ".join(jieba.cut(f))

wordcloud = WordCloud(
    #设置字体，不然会出现口字乱码，文字的路径是电脑的字体一般路径，可以换成别的
    font_path="C:/Windows/Fonts/simfang.ttf",
    #设置了背景，宽高
    background_color="white",width=1000,height=880).generate(cut_text)

plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
