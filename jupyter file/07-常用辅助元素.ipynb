## 1.坐标轴标签、刻度范围和刻度标签

### 1.1坐标轴标签  

xlable(xlable,fontdict=None,labelpad=None)设置x轴标签  
ylabel(ylabel,fontdict=None,labelpad=None)设置y轴标签  
xlabel:表示x轴标签的文本。  
fontdict:表示控制标签文本样式的字典。  
labelpad:表示标签与x轴轴脊间的距离。

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Simhei']
plt.rcParams['axes.unicode_minus'] = False
x=np.linspace(-np.pi,np.pi,256,endpoint=True)  #生成等差间隔数列
y1,y2 = np.sin(x),np.cos(x) 
plt.plot(x,y1,x,y2)
fontdicts = {"family":'Simhei',
            "color":'red',
            "weight":"bold",
            "size":16}
plt.xlabel("x轴",fontdict=fontdicts)
plt.ylabel("y轴",fontdict=fontdicts)
plt.show()
```

### 1.2刻度范围和刻度标签 

## 刻度范围  

xlim(left=None,right=None,emit=True,auto=False,xmin=None,xmax=None)  
left:表示x轴刻度取值区间的左位数。    
right:表示x轴刻度取值区间的右位数。  
emit:表示是否通知限制变化的观察者，默认为True。    
auto:表示是否允许自动缩放x轴，默认为True。     
xmin：表示x轴刻度的最小值。       
xmax：表示x轴刻度的最大值。  

## 刻度标签 

xticks(ticks=None,labels=None)设置x轴刻度标签和刻度位置   
ticks:表示刻度显示的位置列表，该参数可以设置为空列表，以此禁用x轴的刻度。  
labels:表示指定位置刻度的标签列表。   
yticks(ticks=None,labels=None)  

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Simhei']
plt.rcParams['axes.unicode_minus'] = False
x=np.linspace(-np.pi,np.pi,256,endpoint=True)  #生成等差间隔数列
y1,y2 = np.sin(x),np.cos(x) 
plt.plot(x,y1,x,y2)
fontdicts = {"family":'Simhei',
            "color":'red',
            "weight":"bold",
            "size":16}
plt.xlabel("x轴",fontdict=fontdicts)
plt.ylabel("y轴",fontdict=fontdicts)
plt.xlim(x.min()*1.5,x.max()*1.5)
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$',r'$-\pi/2$',r'$0$',r'$\pi/2$',r'$\pi$',])
plt.show()
```

## 2.2添加图例

legend(handles,labels,loc,bbox_to_anchor,ncol,title,shadow,fancybox)  
handles:表示由图形标识构成的列表。  
labels:表示由图例项构成的列表。  
loc:用于控制图例在图表中的位置。  
ncol:表示图例的列数，默认值为1。  
title:表示图例的标题，默认值为None。  
shadow：表示是否在图例后面显示阴影，默认值为None。   
fancybox:表示是否为图例设置圆角边框，默认值为None。  
bbox_to_anchor:用于控制图例的布局，接收一个包含两个数值的元组，第一个数值控制水平位置，数值越大越偏右；第二个数值控制垂直位置，数值越大越往上。  

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Simhei']
plt.rcParams['axes.unicode_minus'] = False
x=np.linspace(-np.pi,np.pi,256,endpoint=True)  #生成等差间隔数列
y1,y2 = np.sin(x),np.cos(x) 
lines=plt.plot(x,y1,x,y2)
fontdicts = {"family":'Simhei',
            "color":'red',
            "weight":"bold",
            "size":16}
plt.xlabel("x轴",fontdict=fontdicts)
plt.ylabel("y轴",fontdict=fontdicts)
plt.xlim(x.min()*1.5,x.max()*1.5)
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$',r'$-\pi/2$',r'$0$',r'$\pi/2$',r'$\pi$',])
plt.title("正弦曲线和余弦曲线",fontdict=fontdicts,loc="left",pad=10)
plt.legend(lines,["正弦","余弦"],shadow=True,fancybox=True,bbox_to_anchor=(1,1))
plt.show()
```

##  4.参考线和参考范围

### 4.1参考线

axhline(y=0,xmax=0,xmin=1,linestyle='-')水平参考线   
y:表示水平参考线的纵坐标    
xmin:水平参考线起始位置 默认0    
xmax:水平参考线的终止位置,默认1    
linestyle:线的类型,默认实线  
axvline(y=0,ymax=0,ymin=1,linestyle='-')垂直参考线   
x:表示水平参考线的横坐标    
ymin:水平参考线起始位置 默认0    
ymax:水平参考线的终止位置,默认1   
linestyle:线的类型,默认实线   

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Simhei']
plt.rcParams['axes.unicode_minus'] = False
x=np.linspace(-np.pi,np.pi,256,endpoint=True)  #生成等差间隔数列
y1,y2 = np.sin(x),np.cos(x) 
lines=plt.plot(x,y1,x,y2)
fontdicts = {"family":'Simhei',
            "color":'red',
            "weight":"bold",
            "size":16}
plt.xlabel("x轴",fontdict=fontdicts)
plt.ylabel("y轴",fontdict=fontdicts)
plt.xlim(x.min()*1.5,x.max()*1.5)
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$',r'$-\pi/2$',r'$0$',r'$\pi/2$',r'$\pi$',])
plt.title("正弦曲线和余弦曲线",fontdict=fontdicts,loc="left",pad=10)
plt.legend(lines,["正弦","余弦"],shadow=True,fancybox=True,bbox_to_anchor=(1,1))
plt.grid(b=True,axis="both",linewidth=0.5)
plt.axhline(y=0,linestyle='-')
plt.axvline(x=0,linestyle='-')
plt.show()
```

### 4.2参考区域

axhspan(ymin,ymax,xmin=0,xmax=1) 水平参考区域 
ymin: 水平跨度的下限,以数据为单位    
ymax: 水平跨度的上限,以数据为单位   
xmin: 垂直跨度的下限,以轴为单位,默认0    
xmax: 垂直跨度的上限,以轴为单位,默认1  
axvspan(ymin,ymax,xmin=0,xmax=1) 垂直参考区域    
xmin: 水平跨度的下限,以数据为单位    
xmax: 水平跨度的上限,以数据为单位   
ymin: 垂直跨度的下限,以轴为单位,默认0
ymax: 垂直跨度的上限,以轴为单位,默认1

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Simhei']
plt.rcParams['axes.unicode_minus'] = False
x=np.linspace(-np.pi,np.pi,256,endpoint=True)  #生成等差间隔数列
y1,y2 = np.sin(x),np.cos(x) 
lines=plt.plot(x,y1,x,y2)
fontdicts = {"family":'Simhei',
            "color":'red',
            "weight":"bold",
            "size":16}
plt.xlabel("x轴",fontdict=fontdicts)
plt.ylabel("y轴",fontdict=fontdicts)
plt.xlim(x.min()*1.5,x.max()*1.5)
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$',r'$-\pi/2$',r'$0$',r'$\pi/2$',r'$\pi$',])
plt.title("正弦曲线和余弦曲线",fontdict=fontdicts,loc="left",pad=10)
plt.legend(lines,["正弦","余弦"],shadow=True,fancybox=True,bbox_to_anchor=(1,1))
plt.grid(b=True,axis="both",linewidth=0.5)
plt.axhline(y=0,linestyle='--')
plt.axvline(x=0,linestyle='--')
plt.axvspan(xmin=0.5,xmax=2,alpha=0.2)
plt.axhspan(ymin=0.5,ymax=1,alpha=0.2)
plt.show()
```

## 5.指向型注释文本和无指向型注释文本  

### 5.1指向型注释文本 

annotate(s,xy)指向型注释文本  
s:注释文本内容  
xy:被注释点的坐标位置,接收元组类型参数(x,y)  
xytext:注释文本的坐标位置(x,y)  
arrowprop:设置箭头的属性字典  
bbox:设置边框的属性字典  


```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Simhei']
plt.rcParams['axes.unicode_minus'] = False
x=np.linspace(-np.pi,np.pi,256,endpoint=True)  #生成等差间隔数列
y1,y2 = np.sin(x),np.cos(x) 
lines=plt.plot(x,y1,x,y2)
fontdicts = {"family":'Simhei',
            "color":'red',
            "weight":"bold",
            "size":16}
plt.xlabel("x轴",fontdict=fontdicts)
plt.ylabel("y轴",fontdict=fontdicts)
plt.xlim(x.min()*1.5,x.max()*1.5)
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$',r'$-\pi/2$',r'$0$',r'$\pi/2$',r'$\pi$',])
plt.title("正弦曲线和余弦曲线",fontdict=fontdicts,loc="left",pad=10)
plt.legend(lines,["正弦","余弦"],shadow=True,fancybox=True,bbox_to_anchor=(1,1))
plt.grid(b=True,axis="both",linewidth=0.5)
plt.axhline(y=0,linestyle='--')
plt.axvline(x=0,linestyle='--')
plt.axvspan(xmin=0.5,xmax=2,alpha=0.2)
plt.axhspan(ymin=0.5,ymax=1,alpha=0.2)
plt.annotate("最小值",
             xy=(-np.pi/2,-1.0),
             xytext=(-(np.pi/2),-0.5),
             arrowprops=dict(arrowstyle="<|-|>")
)
plt.show()
```
