### 绘制条形图或堆积条形图

### 1. 使用barh（）绘制条形图或堆积条形图    
barh（）基础语法：
barh（y，widt=o.8，height，left=None，align=‘center’）  
y:表示条形图的y坐标   
width:表示条形的宽度  
height:表示条形的高度  
left:条形左侧x坐标，默认时0  
align:对齐方式, 'center'和'edge';edge:条形的底边和刻度线对其


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y=np.arange(5)
x1=np.array([10,15,5,8,13])
#设置高度
plt.barh(y,x1,tick_label=['a','b','c','d','e'],height=0.5)
plt.show()
```

### 2. 会绘制多个条形图

```python
y = np.arange(5)
x1 = np.array([10, 15, 5, 8, 13])
x2 = np.array([9, 14, 4, 7, 12])
bar_height = 0.5
# 设置高度
plt.barh(y, x1, tick_label=['a', 'b', 'c', 'd', 'e'], height=bar_height)
plt.barh(y + bar_height, x2, height=bar_height)

plt.show()
```

```python
y = np.arange(5)
x1 = np.array([10, 15, 5, 8, 13])
x2 = np.array([9, 14, 4, 7, 12])
bar_height = 0.5
# 设置高度
plt.barh(y, x1, tick_label=['a', 'b', 'c', 'd', 'e'], height=bar_height)
plt.barh(y,x2,left=x1, height=bar_height)
plt.show()
```

```python
y = np.arange(5)
x1 = np.array([10, 15, 5, 8, 13])
x2 = np.array([9, 14, 4, 7, 12])
error=[2,1,2.5,2,1.5]
bar_height = 0.5
# 设置高度
plt.barh(y, x1, tick_label=['a', 'b', 'c', 'd', 'e'], height=bar_height)
plt.barh(y,x2,left=x1, height=bar_height,xerr=error)
plt.show()
```
